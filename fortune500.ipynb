# Fortune 500

In this notebook, we scrape the [Fortune 500 list](http://beta.fortune.com/fortune500/list).

As it turns out, the webpage grows dinamically as we scroll so, based on [this article](https://blog.scrapinghub.com/2016/06/22/scrapy-tips-from-the-pros-june-2016/), we decided to analyze the network communications done during the scrolling operation (as the page fetches more entries) and found out that it should all be returned by [this url](http://fortune.com/api/v2/list/1666518/expand/item/ranking/asc/0/500). Actually, apparently the server limits to 100 results, so we could for instance fetch 100 at a time (url endings `0/100`, `100/100`, ..., `400/100`).

The amount of data on each company returned by these queries is amazing and, simultaneously, overwhelming.
