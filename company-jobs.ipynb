```python
import pandas as pd
import numpy as np
from geopy.geocoders import Nominatim
import csv
```

```python
companies = ['apple', 'facebook', 'amazon']
```

```python
jobs_df = pd.DataFrame()
for company in companies:
    df = pd.read_csv('data/' + company + '.csv')
    df['company'] = company
    jobs_df = jobs_df.append(df)
```

```python
jobs_df.describe()
```

```python
jobs_df.groupby('company')['company'].agg('count')
```

```python
job_to_locate = jobs_df[['job_id', 'company', 'location']]
# Facebook jobs have more than one location, separated by |, so we should count them as multiple positions
indices = []
new_col = []
for index in xrange(len(job_to_locate)):
    for location in job_to_locate.iloc[index, :]['location'].split('|'):
        new_col.append(location)
        indices.append(index)
job_to_locate = job_to_locate.drop('location')
job_to_locate = job_to_locate.iloc[indices, :]
job_to_locate['location'] = new_col
```

```python
len(jobs_df)
```

```python
len(job_to_locate)
```

```python
geolocator = Nominatim()
# def get_location(location_str):
#     location = geolocator.geocode(location_str)
#     return
locations = job_to_locate[['location']].drop_duplicates()
locations['geolocation'] = locations['location'].apply(geolocator.geocode)
mask = locations['geolocation'].isnull()
locations[mask]
```

```python
locations.loc[mask, 'geolocation'] = locations.loc[mask, 'location']\
    .apply(lambda x: geolocator.geocode(x.replace('Virtual Location - ', '')))
mask = locations['geolocation'].isnull()
locations[mask]
```

```python
locations.loc[mask, 'geolocation'] = locations.loc[mask, 'location']\
    .apply(lambda x: geolocator.geocode(x.replace(', Virtual Contact Center-NC', '')))
mask = locations['geolocation'].isnull()
locations[mask]
```

```python
locations.loc[mask, 'geolocation'] = locations.loc[mask, 'location']\
    .apply(lambda x: geolocator.geocode(x.replace(' (FMK)', '')))
mask = locations['geolocation'].isnull()
locations[mask]
```

```python
locations.loc[mask, 'geolocation'] = locations.loc[mask, 'location']\
    .apply(lambda x: geolocator.geocode(x.replace(' - Corporate', '')))
mask = locations['geolocation'].isnull()
locations[mask]
```

```python
locations.loc[mask, 'geolocation'] = locations.loc[mask, 'location']\
    .apply(lambda x: geolocator.geocode(x.replace(', Virtual Location - Massachuset', '')))
mask = locations['geolocation'].isnull()
locations[mask]
```

```python
job_to_locate = job_to_locate.merge(locations, on='location')
job_to_locate
```

```python
job_to_locate['longitude'] = job_to_locate['geolocation'].apply(lambda x: x.longitude)
job_to_locate['latitude'] = job_to_locate['geolocation'].apply(lambda x: x.latitude)
job_to_locate
```

```python
job_to_locate = job_to_locate.drop('geolocation', axis = 1)
```
